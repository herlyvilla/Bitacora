```python
class Libro:
  def __init__(self, Titulo, Precio ):  #Esta linea se llama constructor
    self.__Titulo = Titulo
    self.__Precio = Precio
  def get_Precio(self): #Para asociar este metodo a la clase se debe usar "self"
    return self.__Precio

  def set_Precio(self, Nuevo_Precio):
    if isinstance(Nuevo_Precio, (int, float)) and Nuevo_Precio > 0:
      self.__Precio = Nuevo_Precio
    else:
      print("El precio debe ser un numero positivo")

  def mostar_informacion(self):
    print(f"Titulo: {self.__Titulo}") #la f indica variable abierta
    print(f"Precio: {self.__Precio:,.2f}") #,. numero de decimales

def main():
  print("----SISTEMA DE LIBROS----")
  Libro1 = Libro("Don Quijote", 120000)
  Libro2 = Libro("Orgullo y prejuicio", 80000)
  print("INFORMACION DEL LIBRO 1")
  print("Los datos del libro 1 son: ")
  Libro1.mostar_informacion()
  print("El precio actual del libro 1 es: ", Libro1.get_Precio())
  print("Modificacion del precio actual: ")
  Libro1.set_Precio(-150000)
  Libro1.set_Precio(150000)
  print("El precio del nuevo libro es:" , Libro1.get_Precio())
  Libro1.mostar_informacion()
  print("----FIN DEL PROGRAMA----")
if __name__ == "__main__":
  main()
```

```python
class Prenda:
  def __init__(self, Nombre, Precio, Stock, Color, Ubicacion ):
    self.__Nombre = Nombre
    self.__Precio = Precio
    self.__Stock = Stock
    self.__Color = Color
    self.__Ubicacion = Ubicacion

  def get_Nombre(self):
    return self.__Nombre

  def get_Precio(self):
    return self.__Precio
  def set_Precio(self, Nuevo_Precio):
    if isinstance(Nuevo_Precio, (int, float)) and Nuevo_Precio > 0:
      self.__Precio = Nuevo_Precio
    else:
      print("El precio debe ser un numero positivo")

  def get_Stock(self):
    return self.__Stock
  def set_Stock(self, Nuevo_Stock):
    if isinstance(Nuevo_Stock, int) and Nuevo_Stock >= 0:
      self.__Stock = Nuevo_Stock
    else:
      print("El stock debe ser un numero entero no negativo")

  def get_Color(self):
    return self.__Color

  def get_Ubicacion(self):
    return self.__Ubicacion

  # Metodo para mostrar informacion
  def mostar_informacion(self):
    print(f"Nombre: {self.__Nombre}")
    print(f"Precio: {self.__Precio:,.2f}")
    print(f"Stock: {self.__Stock}")
    print(f"Color: {self.__Color}")
    print(f"Ubicacion: {self.__Ubicacion}")

prenda1 = Prenda("Camiseta", 25.50, 100, "Azul", "Estante A")
prenda2 = Prenda("Pantalón", 35.00, 50, "Negro", "Estante B")

print("----DATOS DE PRENDAS----")
print()
print("INFORMACION PRENDA 1")
print()
print(f"Nombre: {prenda1.get_Nombre()}")
print(f"Precio:{prenda1.get_Precio()}")
print(f"Stock:{prenda1.get_Stock()}")
print(f"Color:{prenda1.get_Color()}")
print(f"Ubicacion:{prenda1.get_Ubicacion()}")
print()
print("INFORMACION PRENDA 2")
print()
print(f"Nombre: {prenda2.get_Nombre()}")
print(f"Precio:{prenda2.get_Precio()}")
print(f"Stock:{prenda2.get_Stock()}")
print(f"Color:{prenda2.get_Color()}")
print(f"Ubicacion:{prenda2.get_Ubicacion()}")
print()
print("----FIN DEL PROGRAMA----\n")

#Modificacion del usuario
opcion = input("Ingrese el nombre de la prenda que desea modificar: ")

if opcion == prenda1.get_Nombre():
    nuevo_precio_str = input(f"\nIngrese el nuevo precio para {prenda1.get_Nombre()}: ")
    try:
        nuevo_precio = float(nuevo_precio_str)
        prenda1.set_Precio(nuevo_precio)
        print()
        print(f"Precio de {prenda1.get_Nombre()} actualizado a {prenda1.get_Precio():,.2f}")
        print("Los nuevos datos de la prenda son:")
        print()
        prenda1.mostar_informacion() #como puedo mostar todo de nuevo con el cambio?
    except ValueError:
        print("Entrada inválida. Por favor, ingrese un número para el precio.")
elif opcion == prenda2.get_Nombre():
    nuevo_precio_str = input(f"\nIngrese el nuevo precio para {prenda2.get_Nombre()}: ")
    try:
        nuevo_precio = float(nuevo_precio_str)
        prenda2.set_Precio(nuevo_precio)
        print()
        print(f"Precio de {prenda2.get_Nombre()} actualizado a {prenda2.get_Precio():,.2f}")
    except ValueError:
        print("Entrada inválida. Por favor, ingrese un número para el precio.")
else:
    print("Prenda no encontrada")

print("\n----FIN DEL PROGRAMA----")
```
